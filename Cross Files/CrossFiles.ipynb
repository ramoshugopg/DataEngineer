```python
import pandas as pd
import numpy as np
```

```python
#Convert all to a data frame
df = pd.read_csv("Crosswalk file update, 4-25(Text).csv", encoding="utf-8")
df.columns
```

```python
#Filter to get the rigth schema
columns_to_keep = ['File ID', 'Respondent ID','S1','AGE_RANGE','S2','S4','S5_BRAND1', 'S5_BRAND2', 
'S5_BRAND3','S5_BRAND4','S5_BRAND5','S5_BRAND6','S6','S7_BRAND1','S7_BRAND2','S7_BRAND3',
'S7_BRAND4','S8_BRAND1','S8_BRAND2','S8_BRAND3','S8_BRAND4','S9','S10', 'S11', 'S12']
filtered_df = df[columns_to_keep]

filtered_df.head(5)
```

```python
#Start transformation of the columns S5_BRAND to boolean
filtered_df['S5_BRAND1'] = filtered_df['S5_BRAND1'].apply(lambda x: True if x == 'YouTube' else False)
filtered_df['S5_BRAND2'] = filtered_df['S5_BRAND2'].apply(lambda x: True if x == 'Facebook' else False)
filtered_df['S5_BRAND3'] = filtered_df['S5_BRAND3'].apply(lambda x: True if x == 'Instagram' else False)
filtered_df['S5_BRAND4'] = filtered_df['S5_BRAND4'].apply(lambda x: True if x == 'TikTok' else False)
filtered_df['S5_BRAND5'] = filtered_df['S5_BRAND5'].apply(lambda x: True if x == 'Google search' else False)
filtered_df['S5_BRAND6'] = filtered_df['S5_BRAND6'].apply(lambda x: True if x == 'Amazon' else False)
```

```python
#Convert S7 and S8 brands into integers with clusters
# Define your mapping
time_mapping = {
    "Less than 30 minutes": 1,
    "30 minutes to 1 hour": 2,
    "1 to 2 hours": 3,
    "2 to 4 hours": 4,
    "4 to 6 hours": 5,
    "Over 6 hours": 6
}

# Apply mapping to create a new integer column
filtered_df['S7_BRAND1'] = filtered_df['S7_BRAND1'].map(time_mapping)
filtered_df['S7_BRAND2'] = filtered_df['S7_BRAND2'].map(time_mapping)
filtered_df['S7_BRAND3'] = filtered_df['S7_BRAND3'].map(time_mapping)
filtered_df['S7_BRAND4'] = filtered_df['S7_BRAND4'].map(time_mapping)
filtered_df['S8_BRAND1'] = filtered_df['S8_BRAND1'].map(time_mapping)
filtered_df['S8_BRAND2'] = filtered_df['S8_BRAND2'].map(time_mapping)
filtered_df['S8_BRAND3'] = filtered_df['S8_BRAND3'].map(time_mapping)
filtered_df['S8_BRAND4'] = filtered_df['S8_BRAND4'].map(time_mapping)
```
