```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/hugoramos/Downloads/lumen_Lumen norms for P&G - June 2025.csv')
df
```

```python
dm = pd.read_excel('/Users/hugoramos/Downloads/lumen_media_mapping.xlsx')
dm.columns
```

```python
#Create keys for the column mapping
keys = list(zip(dm['channel'],dm['platform'],dm['media_type']))

#Create mapping dictionaries
media_subtype_map = dict(zip(keys, dm['media_subtype']))
app_type_map = dict(zip(keys, dm['app_type']))
publisher_map = dict(zip(keys, dm['publisher']))
```

```python
publisher_map
```

```python
def replace_nan_with_none(key):
    return tuple(None if pd.isna(k) else k for k in key)

media_subtype_map = {
    replace_nan_with_none(k): v
    for k, v in media_subtype_map.items()
}
app_type_map = {
    replace_nan_with_none(k): v
    for k, v in app_type_map.items()
}
publisher_map = {
    replace_nan_with_none(k): v
    for k, v in publisher_map.items()
}
```

```python
# Create all the column by mapping each row
df["media_subtype"] = df.apply(
    lambda row: media_subtype_map.get((row["channel"],row["platform"],row["media_type"])),
    axis=1)

df["app_type"] = df.apply(
    lambda row: app_type_map.get((row["channel"],row["platform"],row["media_type"])),
    axis=1)

df["publisher"] = df.apply(
    lambda row: publisher_map.get((row["channel"],row["platform"],row["media_type"])),
    axis=1)
```
